### 2) Corelation b/w Features

```python
#corelation matrix
# Numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=[np.number])

# correlation matrix
corr_matrix = numeric_df.corr()

# Create the mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(30, 12))

# Draw the heatmap
sns.heatmap(corr_matrix, mask=mask, square=True, cbar=True, ax=ax, annot=corr_matrix,  annot_kws={"size": 10, "color": "black"}, cmap="coolwarm" )

# Show the plot
plt.show()
```

SOME INFERENCES FROM THE ABOVE HEATMAP
Self relation ie of a feature to itself is equal to 1 as expected.

JobLevel is highly related to Age as expected as aged employees will generally tend to occupy higher positions in the company.

MonthlyIncome is very strongly related to joblevel as expected as senior employees will definately earn more.

PerformanceRating is highly related to PercentSalaryHike which is quite obvious.

Also note that TotalWorkingYears is highly related to JobLevel which is expected as senior employees must have worked for a larger span of time.

YearsWithCurrManager is highly related to YearsAtCompany.

YearsAtCompany is related to YearsInCurrentRole.



### 3) Feature Selection

### 3.1) Plotting the Feature against the 'Target' variable

#### 3.1.1) Employee Attrition

```python
attrition_yes = df['Attrition'] == 'Yes'
attrition_yes_count = attrition_yes.sum()
attrition_total = attrition_yes.count()
attrition_yes_percentage = (attrition_yes_count * 100)/attrition_total


attrition_no = df['Attrition'] == 'No'
attrition_no_count = attrition_no.sum()
attrition_no_percentage = (attrition_no_count * 100)/attrition_total


print("Attrition Percentage of peoples in the organization : ", attrition_yes_percentage)
print("Percentage of peoples in the organization: ", attrition_no_percentage)
```

From the given data, overall 84% of employees wants to stay in the organization and 16% wants to leave the organization.

#### 3.1.2) Gender Analysis

```python
male_attrition = ((df['Attrition'] == 'Yes') & (df['Gender'] == 'Male')).sum()
total_male = (df['Gender'] == 'Male').sum()
male_attrition_percentage = (male_attrition * 100)/ total_male

female_attrition = ((df['Attrition'] == 'Yes') & (df['Gender'] == 'Female')).sum()
total_female = (df['Gender'] == 'Female').sum()
female_attrition_percentage = (female_attrition * 100)/ total_female

print("Percentage of male attrition:", male_attrition_percentage)
print("Percentage of female attrition:", female_attrition_percentage)
```

About 85% of female employees wants to stay in the organization while only 15% want to leave the organization. However, about 83% of male employees want to be in the organization with only 17% wants to leave the organization.

#### 3.1.3) Age Analysis

Age is a continuous quantity and therefore can plot it against the Attrition using a boxplot. 

```python
sns.catplot(data=df, y='Age', x='Attrition', kind='box', aspect=1, height=5)
```

The median as well as the maximum age of the people with 'No' attrition is higher than that of 'Yes' category. This represents that peoples with higher age group have lesser tendency to leave the organization which makes sense as they may have settled in the organization.

#### 3.1.4) Department Analysis

```python
df['Department'].unique()
```

```python
sns.catplot(x='Attrition', data=df, kind='count', col='Department', aspect=1)
```

```python
sales_attrition = ((df['Department'] == 'Sales') & (df['Attrition'] == 'Yes')).sum()
rd_attrition = ((df['Department'] == 'Research & Development') & (df['Attrition'] == 'Yes')).sum()
hr_attrition = ((df['Department'] == 'Human Resources') & (df['Attrition'] == 'Yes')).sum()

total_sales = (df['Department'] == 'Sales').sum()
total_rd = (df['Department'] == 'Research & Development').sum()
total_hr = (df['Department'] == 'Human Resources').sum()

sales_attrition_percentage = (sales_attrition * 100)/total_sales
rd_attrition_percentage = (rd_attrition * 100)/total_rd
hr_attrition_percentage = (hr_attrition * 100)/total_hr

print("Sales Attrition Percentage:", sales_attrition_percentage)
print("Research and Development Attrition Percentage:", rd_attrition_percentage)
print("Human Resource Attrition Percentage:", hr_attrition_percentage) 
```

#### 3.1.5) Job Level

```python
df['JobLevel'].unique()
```

```python
attrition_j1 = ((df['JobLevel'] == 1) & (df['Attrition'] == 'Yes')).sum()
attrition_j2 = ((df['JobLevel'] == 2) & (df['Attrition'] == 'Yes')).sum()
attrition_j3 = ((df['JobLevel'] == 3) & (df['Attrition'] == 'Yes')).sum()
attrition_j4 = ((df['JobLevel'] == 4) & (df['Attrition'] == 'Yes')).sum()
attrition_j5 = ((df['JobLevel'] == 5) & (df['Attrition'] == 'Yes')).sum()

total_attrition_j1 = (df['JobLevel'] == 1).sum()
total_attrition_j2 = (df['JobLevel'] == 2).sum()
total_attrition_j3 = (df['JobLevel'] == 3).sum()
total_attrition_j4 = (df['JobLevel'] == 4).sum()
total_attrition_j5 = (df['JobLevel'] == 5).sum()

attrition_j1_percentage = (attrition_j1 * 100)/total_attrition_j1
attrition_j2_percentage = (attrition_j2 * 100)/total_attrition_j2
attrition_j3_percentage = (attrition_j3 * 100)/total_attrition_j3
attrition_j4_percentage = (attrition_j4 * 100)/total_attrition_j4
attrition_j5_percentage = (attrition_j5 * 100)/total_attrition_j5

print("Attrition Percentage of Job Level 1:", attrition_j1_percentage)
print("Attrition Percentage of Job Level 2:", attrition_j2_percentage)
print("Attrition Percentage of Job Level 3:", attrition_j3_percentage)
print("Attrition Percentage of Job Level 4:", attrition_j4_percentage)
print("Attrition Percentage of Job Level 5:", attrition_j5_percentage)
```

People in JobLevel 4 have less Attrition percentage and they are more likely to stay in the organization.  Similar inferences can be made for other job levels.
